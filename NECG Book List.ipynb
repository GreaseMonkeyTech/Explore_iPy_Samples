```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/LuisFRoch/Datasets/master/NECG_Book-List.csv')

fig = go.Figure(data=[go.Table(
    header=dict(values=list(df.columns),
                fill_color='royalblue',
                line_color='darkslategray',
                font=dict(color='white', size=12),
                align='left'),
    cells=dict(values=[df.Title, df.Author],
               line_color='darkslategray',
               fill_color='white',
               align='left'))
])

fig.update_layout(
    height=900,
    title_text="NECG Book List",
)

fig.show()
```
