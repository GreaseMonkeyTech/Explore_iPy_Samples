```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
import seaborn as sns
import math
import plotly
from plotly.graph_objs import Scatter, Layout
plotly.offline.init_notebook_mode(connected=True)
```

```python
# TedSpread -monthly end of period- continuously compounded rate of change
ted = pd.read_csv('documents/mission_v.01/TEDRATE.csv')

# CBOE Volatility Index -monthly end of period- continuously compounded rate of change
vix = pd.read_csv('documents/mission_v.01/VIXCLS.csv')

# 5-Year High Quality Market (HQM) Corporate Bond Par Yield -monthly- continuously compounded rate of change
hqm5 = pd.read_csv('documents/mission_v.01/HQMCB5YRP.csv')

# US Corporate AAA Option-Adjusted Spread -monthly end of period- continuously compounded rate of change
c3a = pd.read_csv('documents/mission_v.01/BAMLC0A1CAAA.csv')
```

```python
# Merging Data-files by "Date":
# df1 = pd.merge(vix,hqm5,on='DATE').merge(ted,on='DATE').merge(c3a,on='DATE')

df = ted.merge(vix, on = 'DATE')
df1 = df.merge(hqm5, on = 'DATE')
df2 = df1.merge(c3a, on = 'DATE')
```

```python
# Format data type to "Date-time". Show data frame 

pd.to_datetime(df2['DATE'])
df2.head()
```

```python
# Review distributions

from pandas.tools.plotting import scatter_matrix
scatter_matrix(df2, alpha=0.8, figsize=(30, 15), diagonal='kde')
```

```python
# Compare data distributions
df2.plot(kind='kde', figsize=(15, 10))
```

```python
# Histogram

df2.plot(kind='hist', figsize=(15, 10), bins=30)
```

```python
# Simple line graph

df2.plot(x='DATE', kind='area', stacked=False, figsize=(15,10));
```

```python
# Boxplot to obsserve outliers

df2.plot(kind='box', stacked=False, figsize=(15,10))
```
